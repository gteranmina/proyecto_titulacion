```python
import pandas as pd
import numpy as np
import re
import os

# Definir el nombre base de los archivos
id_file = "Data/id_ecuador_complete.csv"
tax_file_template = "Data/totaltax_ND({}).csv"

# Lista de años a procesar (2007 a 2015)
years = list(range(2007, 2016))

print(f"Años a procesar: {years}")

for year in years:
    # Crear la carpeta si no existe
    output_folder = str(year)
    os.makedirs(output_folder, exist_ok=True)
    
    tax_file = tax_file_template.format(year)
    
    # Cargar los archivos CSV
    id_ecuador = pd.read_csv(id_file)
    total_tax = pd.read_csv(tax_file)
    
    # Unir los DataFrames usando el identificador único sin sobreponer columnas
    merged_df = total_tax.merge(id_ecuador, left_on="id_inf", right_on="id", how="left")
    
    # Eliminar la columna duplicada de id
    merged_df.drop(columns=["id"], inplace=True)
    
    # Eliminar el valor 99999 de ciiu4n4
    merged_df = merged_df[merged_df['ciiu4n4'] != '99999']
    # imprime cuantos valores 99999 se eliminaron
    print(f"Para el año {year} se eliminaron las filas con ciiu4n4 = 99999")
    
    # Asignar el año correspondiente a toda la columna 'anio'
    merged_df['anio'] = year
    
    # Guardar el resultado en un nuevo archivo CSV
    output_file = os.path.join(output_folder, f"merged_data_ecuador_{year}.csv")
    merged_df.to_csv(output_file, index=False)
    
    print(f"Archivo '{output_file}' creado exitosamente con la columna 'anio' llena con {year}.")
```

# Matriz de canton-industria-total_tax para cada año

```python
import pandas as pd
import numpy as np
import os

# Lista de años a procesar (2007 a 2015)
years = list(range(2007, 2016))

print(f"Años a procesar: {years}")

for year in years:
    # Crear la carpeta si no existe
    output_folder = str(year)
    os.makedirs(output_folder, exist_ok=True)
    
    # Cargar el archivo mergeado
    merged_file = os.path.join(output_folder, f"merged_new_data_ecuador_{year}.csv")
    df = pd.read_csv(merged_file)
    
    # Asegurar que los datos sean del tipo correcto y filtrar valores inesperados
    df = df[df['ciiu4n4'].notna() & df['city_code'].notna()]
    df['ciiu4n4'] = df['ciiu4n4'].astype(str).str.strip()
    df['city_code'] = df['city_code'].astype(str).str.strip()
    
    # Crear la matriz pivotante asegurando que las intersecciones sean valores de total_tax
    tax_matrix = df.pivot_table(index='city_code', columns='ciiu4n4', values='total_tax', aggfunc=np.sum, fill_value=0)
    
    # Guardar la matriz en un archivo CSV
    output_matrix_file = os.path.join(output_folder, f"tax_matrix_{year}.csv")
    tax_matrix.to_csv(output_matrix_file, index=True)
    
    print(f"Matriz de impuestos para {year} generada y guardada en '{output_matrix_file}'.")
```

# Calculo de RCA y Relatedness

```python
import os
import pandas as pd
import numpy as np
import economic_complexity as ec

# Lista de años a procesar (2007 a 2015)
years = list(range(2007, 2016))

print(f"Años a procesar: {years}")

for year in years:
    # Crear la carpeta si no existe
    output_folder = str(year)
    os.makedirs(output_folder, exist_ok=True)

    # Cargar datos
    df = pd.read_csv(os.path.join(output_folder, f"merged_new_data_ecuador_{year}.csv"))

    # Crear copias temporales sin modificar las columnas originales
    df_temp = df[['city_code', 'ciiu4n4', 'total_tax']].copy()
    
    # Renombrar columnas en la copia para la manipulación sin afectar el original
    df_temp = df_temp.rename(columns={'city_code': 'location', 'ciiu4n4': 'activity', 'total_tax': 'value'})

    # Crear una tabla pivotante con ubicaciones como filas y actividades como columnas
    pivot_exports = df_temp.pivot_table(
        values="value",
        index="location",
        columns="activity",
        aggfunc=np.sum,
        fill_value=0  # Para evitar NaN
    )

    # Calcular RCA
    rca_matrix = ec.rca(pivot_exports)

    # Convertir la matriz en un DataFrame binario (1 si RCA >= 1, 0 en caso contrario)
    df_rca_binary = pd.DataFrame((rca_matrix >= 1).astype(int), index=pivot_exports.index, columns=pivot_exports.columns)

    # Guardar matriz de RCA binaria
    rca_binary_file = os.path.join(output_folder, f"rca_matrix_binary_{year}.csv")
    df_rca_binary.to_csv(rca_binary_file)
    print(f"Matriz de RCA Binaria para {year} guardada en '{rca_binary_file}'.")

    # Calcular la matriz de proximidad entre actividades económicas (Relatedness en valores continuos)
    relatedness_matrix = ec.proximity(rca_matrix)  # Usamos la matriz RCA original (no binaria)

    # Guardar matriz de Relatedness continua
    df_relatedness = pd.DataFrame(relatedness_matrix, index=pivot_exports.columns, columns=pivot_exports.columns)
    relatedness_file = os.path.join(output_folder, f"relatedness_matrix_{year}.csv")
    df_relatedness.to_csv(relatedness_file)
    print(f"Matriz de Relatedness (valores continuos) para {year} guardada en '{relatedness_file}'.")

    # Mostrar primeros resultados
    print(f"Matriz de RCA binaria para {year}")
    print(df_rca_binary.head())

    print(f"Matriz de Relatedness continua para {year}")
    print(df_relatedness.head())
```

# Matriz de comparacion de productos nuevos entre cada año

```python
import pandas as pd
import os
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import StandardScaler
import umap.umap_ as umap

# Lista de años a procesar (2007 a 2015)
years = list(range(2007, 2016))
print(f"Años a procesar: {years}")

# Carpeta para guardar las matrices de nuevos productos originales
original_folder = "new_products_matrices"
os.makedirs(original_folder, exist_ok=True)

# Carpeta para almacenar los datos preprocesados (diferenciados de los originales)
preprocessed_folder = "preprocessed_new_products"
os.makedirs(preprocessed_folder, exist_ok=True)

# Parámetros de preprocesamiento
variance_threshold = 0.01   # Elimina columnas con varianza menor a este valor
umap_components = 10        # Número de componentes para UMAP

# Procesar año a año (comparación entre año anterior y actual)
for i in range(len(years) - 1):
    year_prev = years[i]
    year_current = years[i + 1]
    
    # Rutas de los archivos CSV para cada año
    file_prev = os.path.join(str(year_prev), f"rca_matrix_binary_{year_prev}.csv")
    file_current = os.path.join(str(year_current), f"rca_matrix_binary_{year_current}.csv")
    
    # Verificar que ambos archivos existen
    if not os.path.exists(file_prev) or not os.path.exists(file_current):
        print(f"Archivos no encontrados: {file_prev} o {file_current}. Saltando comparación {year_prev}-{year_current}.")
        continue

    # Cargar datos con el índice correspondiente (código único de ciudad)
    df_prev = pd.read_csv(file_prev, index_col=0)
    df_current = pd.read_csv(file_current, index_col=0)
    
    # Alinear índices y columnas para asegurar la comparabilidad
    df_prev, df_current = df_prev.align(df_current, fill_value=0)
    
    # Generar la matriz de nuevos productos:
    # Se asigna 1 si el producto aparece en el año actual y no estaba en el año anterior, 0 en caso contrario.
    new_products = ((df_current == 1) & (df_prev == 0)).astype(int)
    
    # Guardar la matriz original de nuevos productos
    original_file = os.path.join(original_folder, f"new_products_matrix_{year_prev}_{year_current}.csv")
    new_products.to_csv(original_file)
    print(f"Matriz original para {year_prev}-{year_current} guardada en '{original_file}'.")
    
    # --- Pipeline de preprocesamiento ---
    # 1. Eliminación de variables casi constantes
    vt = VarianceThreshold(threshold=variance_threshold)
    new_products_vt = vt.fit_transform(new_products)
    print(f"{year_prev}-{year_current}: Dimensión original {new_products.shape}, tras VarianceThreshold {new_products_vt.shape}")

    # 2. Escalado de características
    scaler = StandardScaler()
    new_products_scaled = scaler.fit_transform(new_products_vt)
    
    # 3. Reducción de dimensionalidad con UMAP para preservar la estructura de "relatedness density"
    reducer = umap.UMAP(n_components=umap_components, random_state=42)
    new_products_umap = reducer.fit_transform(new_products_scaled)
    
    # Convertir la matriz reducida a DataFrame, manteniendo el índice original (código único de ciudad)
    preprocessed_df = pd.DataFrame(new_products_umap, index=new_products.index,
                                   columns=[f"umap_{j+1}" for j in range(umap_components)])
    
    # Guardar los datos preprocesados en la carpeta designada, con un nombre distintivo
    preprocessed_file = os.path.join(preprocessed_folder, f"preprocessed_new_products_matrix_{year_prev}_{year_current}.csv")
    preprocessed_df.to_csv(preprocessed_file)
    print(f"Datos preprocesados para {year_prev}-{year_current} guardados en '{preprocessed_file}'.\n")

print("Proceso finalizado. Se han generado y guardado tanto las matrices originales de nuevos productos como sus versiones preprocesadas.")
```

# Matriz de comparacion de productos desaparecidos entre cada año y Preprocesamiento de datos

```python
import pandas as pd
import os

# Lista de años a procesar (2007 a 2015)
years = list(range(2007, 2016))

print(f"Años a procesar: {years}")

# Carpeta para almacenar las matrices de productos desaparecidos
disappeared_products_folder = "disappeared_products_matrices"
os.makedirs(disappeared_products_folder, exist_ok=True)

# Diccionario para almacenar las matrices de productos desaparecidos
disappeared_product_matrices = {}

for i in range(len(years) - 1):
    year_current = years[i + 1]  # Año actual (ej. 2015)
    year_prev = years[i]  # Año anterior (ej. 2014)

    # Definir rutas de archivos de entrada
    file_current = os.path.join(str(year_current), f"rca_matrix_binary_{year_current}.csv")
    file_prev = os.path.join(str(year_prev), f"rca_matrix_binary_{year_prev}.csv")

    # Verificar si ambos archivos existen antes de continuar
    if not os.path.exists(file_current) or not os.path.exists(file_prev):
        print(f"Archivos no encontrados: {file_prev} o {file_current}. Saltando comparación {year_prev}-{year_current}.")
        continue

    # Cargar datos de ambos años
    df_current = pd.read_csv(file_current, index_col=0)
    df_prev = pd.read_csv(file_prev, index_col=0)

    # Asegurar que ambas matrices tengan el mismo índice y columnas
    df_prev, df_current = df_prev.align(df_current, fill_value=0)

    # **Matriz de productos desaparecidos (-1 si desapareció, 0 si sigue existiendo o no existía)**
    disappeared_products = ((df_current == 0) & (df_prev == 1)).astype(int) * -1
    disappeared_product_matrices[f"{year_prev}_{year_current}"] = disappeared_products

    # **Guardar la matriz de productos desaparecidos**
    disappeared_products_file = os.path.join(disappeared_products_folder, f"disappeared_products_matrix_{year_prev}_{year_current}.csv")
    disappeared_products.to_csv(disappeared_products_file)
    print(f"Matriz de productos desaparecidos para {year_prev}-{year_current} guardada en '{disappeared_products_file}'.")

print("Proceso finalizado. Todas las matrices de productos desaparecidos han sido generadas.")
```

# Código para entrenar XGBoost (demasiado Overfitting)

Este código:

- Lee las matrices de productos nuevos.
- Entrena XGBoost para predecir productos nuevos.
- Calcula las métricas: BestF1, Precision@1000, mAP@20.
- Guarda los modelos entrenados.

# Opción de Ventana deslizante

Para cada paso, entrenas con la concatenación de las ventanas anteriores y pruebas con la ventana siguiente. Se realiza un reindexado (alineación de columnas) en cada iteración para evitar la discrepancia de características.

```python
import pandas as pd
import os
import numpy as np
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, precision_score, average_precision_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import joblib

os.makedirs('trained_models', exist_ok=True)

def optimize_threshold(y_true, y_probs):
    thresholds = np.linspace(0.1, 0.9, 81)
    best_thresh, best_f1 = 0.5, 0
    for t in thresholds:
        current_f1 = f1_score(y_true, (y_probs[:, 1] >= t).astype(int))
        if current_f1 > best_f1:
            best_f1, best_thresh = current_f1, t
    return best_thresh, best_f1

data_folder = "new_products_matrices"
files = sorted([f for f in os.listdir(data_folder) if f.startswith("new_products_matrix_") and f.endswith(".csv")])

window_size = 1

results = []

for i in range(len(files) - window_size):
    print(files)
    train_dfs = [pd.read_csv(os.path.join(data_folder, files[j]), index_col=0) for j in range(i, i + window_size)]
    train_data = pd.concat(train_dfs, axis=0)
    test_data = pd.read_csv(os.path.join(data_folder, files[i + window_size]), index_col=0)

    all_columns = sorted(set(train_data.columns).union(test_data.columns))
    train_data = train_data.reindex(columns=all_columns, fill_value=0)
    test_data = test_data.reindex(columns=all_columns, fill_value=0)

    X_train, y_train = train_data.values, (train_data.sum(axis=1) > 0).astype(int)
    X_test, y_test = test_data.values, (test_data.sum(axis=1) > 0).astype(int)

    param_grid = {
        'max_depth': [3, 4, 5],
        'reg_alpha': [1, 2, 5],
        'reg_lambda': [10, 20, 50]
    }

    xgb = XGBClassifier(
        eval_metric="logloss",
        learning_rate=0.05,
        n_estimators=200,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

    clf = GridSearchCV(xgb, param_grid, scoring='f1', cv=3)
    clf.fit(X_train, y_train)

    best_model = clf.best_estimator_

    eval_set = [(X_train, y_train), (X_test, y_test)]
    best_model.fit(
        X_train, y_train,
        eval_set=eval_set,
        verbose=False
    )

    results_eval = best_model.evals_result()

    plt.figure()
    plt.plot(results_eval['validation_0']['logloss'], label='train')
    plt.plot(results_eval['validation_1']['logloss'], label='test')
    plt.xlabel('Iterations')
    plt.ylabel('LogLoss')
    plt.title(f'LogLoss Curves {files[i + window_size]} \n XGBoost sin Oversampling y con datos originales')
    plt.legend()
    plt.show()

    y_probs = best_model.predict_proba(X_test)
    best_thresh, best_f1 = optimize_threshold(y_test, y_probs)
    y_pred_opt = (y_probs[:, 1] >= best_thresh).astype(int)

    f1_opt = f1_score(y_test, y_pred_opt)
    prec_opt = precision_score(y_test, y_pred_opt)
    ap20_opt = average_precision_score(y_test, y_pred_opt)

    results.append((files[i + window_size], best_thresh, f1_opt, prec_opt, ap20_opt))

    print(f"Ventana Test: {files[i + window_size]}")
    print(f"  Umbral óptimo: {best_thresh:.2f}")
    print(f"  F1: {f1_opt:.2f} | Precision@1000: {prec_opt:.2f} | mAP@20: {ap20_opt:.2f}\n")

    joblib.dump(best_model, f'trained_models/best_model_{files[i + window_size]}.model')

avg_f1 = np.mean([r[2] for r in results])
avg_prec = np.mean([r[3] for r in results])
avg_ap20 = np.mean([r[4] for r in results])

print("Resumen global:")
print(f"  F1 promedio: {avg_f1:.2f}")
print(f"  Precision promedio: {avg_prec:.2f}")
print(f"  mAP@20 promedio: {avg_ap20:.2f}")
```

# Código para entrenar ElasticNet logistic regression

```python
import pandas as pd
import numpy as np
import os
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import f1_score, precision_score, average_precision_score
import joblib
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

os.makedirs('trained_models', exist_ok=True)

data_folder = "new_products_matrices"
files = sorted([f for f in os.listdir(data_folder) if f.startswith("new_products_matrix_") and f.endswith(".csv")])

window_size = 1
results_logreg = []

def optimize_threshold(y_true, y_probs):
    thresholds = np.linspace(0.1, 0.9, 81)
    best_thresh, best_f1 = 0.5, 0
    for t in thresholds:
        current_f1 = f1_score(y_true, (y_probs[:, 1] >= t).astype(int))
        if current_f1 > best_f1:
            best_f1, best_thresh = current_f1, t
    return best_thresh, best_f1

imputer = SimpleImputer(strategy='mean')

for i in range(len(files) - window_size):
    train_dfs = [pd.read_csv(os.path.join(data_folder, files[j]), index_col=0) for j in range(i, i + window_size)]
    train_data = pd.concat(train_dfs, axis=0)
    test_data = pd.read_csv(os.path.join(data_folder, files[i + window_size]), index_col=0)

    all_columns = sorted(set(train_data.columns).union(test_data.columns))
    train_data = train_data.reindex(columns=all_columns, fill_value=0)
    test_data = test_data.reindex(columns=all_columns, fill_value=0)

    X_train, y_train = train_data.values, (train_data.sum(axis=1) > 0).astype(int)
    X_test, y_test = test_data.values, (test_data.sum(axis=1) > 0).astype(int)

    # Imputar NaN con la media
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)
    
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    model = LogisticRegressionCV(
        cv=3,
        penalty='elasticnet',
        solver='saga',
        l1_ratios=[0.1, 0.5, 0.9],
        scoring='f1',
        random_state=42,
        max_iter=2000,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    y_probs = model.predict_proba(X_test)
    best_thresh, best_f1 = optimize_threshold(y_test, y_probs)
    y_pred_opt = (y_probs[:, 1] >= best_thresh).astype(int)

    f1_opt = f1_score(y_test, y_pred_opt)
    prec_opt = precision_score(y_test, y_pred_opt)
    ap20_opt = average_precision_score(y_test, y_pred_opt)

    results_logreg.append((files[i + window_size], best_thresh, f1_opt, prec_opt, ap20_opt))

    print(f"LogReg - Ventana Test: {files[i + window_size]}")
    print(f"  Umbral óptimo: {best_thresh:.2f}")
    print(f"  F1: {f1_opt:.2f} | Precision: {prec_opt:.2f} | mAP@20: {ap20_opt:.2f}\n")

    joblib.dump(model, f'trained_models/logreg_{files[i + window_size]}.model')
```

```python
import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import f1_score, precision_score, average_precision_score, classification_report
from sklearn.preprocessing import StandardScaler

import xgboost as xgb
from xgboost import XGBClassifier

os.makedirs('trained_models', exist_ok=True)

# --- Configuración de rutas y carga de datos ---
preproc_folder = "preprocessed_new_products"  # Carpeta con los datos preprocesados (features UMAP)
orig_folder = "new_products_matrices"         # Carpeta con las matrices originales (para definir el target)

# Inicializar listas para almacenar features y targets de cada período
list_features = []
list_targets = []

# Usar datos de 2007 a 2015:
# - Features: desde 2007-2008 hasta 2013-2014
# - Target: desde 2008-2009 hasta 2014-2015
for year in range(2007, 2014):  # años: 2007, 2008, ..., 2013
    features_file = os.path.join(preproc_folder, f"preprocessed_new_products_matrix_{year}_{year+1}.csv")
    target_file   = os.path.join(orig_folder, f"new_products_matrix_{year+1}_{year+2}.csv")
    
    # Cargar features y target para el período
    df_features = pd.read_csv(features_file, index_col=0)
    df_target = pd.read_csv(target_file, index_col=0)
    
    # Definir el target: 1 si la suma de la fila (productos nuevos) es > 0, 0 en caso contrario.
    y_period = (df_target.sum(axis=1) > 0).astype(int)
    
    # Alinear índices entre features y target
    df_features, y_period = df_features.align(y_period, axis=0, join='inner')
    
    # Agregar a las listas
    list_features.append(df_features)
    list_targets.append(y_period)

# Concatenar todos los dataframes de features y targets
df_features_all = pd.concat(list_features, axis=0)
y_all = pd.concat(list_targets, axis=0)

# Convertir a arrays para el modelo
X = df_features_all.values
y = y_all.values

# --- Preprocesamiento ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en conjuntos de entrenamiento y prueba (80% / 20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# --- Ajuste de hiperparámetros con RandomizedSearchCV ---
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 1.5, 2]
}

clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
random_search = RandomizedSearchCV(
    clf, 
    param_distributions=param_grid, 
    scoring='f1', 
    cv=3, 
    n_iter=20, 
    n_jobs=-1, 
    random_state=42
)
random_search.fit(X_train, y_train)
best_model = random_search.best_estimator_
print("Mejores hiperparámetros encontrados:", random_search.best_params_)

# --- Entrenamiento final ---
eval_set = [(X_train, y_train), (X_test, y_test)]
best_model.fit(
    X_train, y_train, 
    eval_set=eval_set, 
    verbose=True
)

# Obtener el historial de evaluación
evals_result = best_model.evals_result()

# --- Curvas de entrenamiento y validación ---
epochs = len(evals_result['validation_0']['logloss'])
x_axis = range(epochs)
plt.figure(figsize=(10, 6))
plt.plot(x_axis, evals_result['validation_0']['logloss'], label='Entrenamiento')
plt.plot(x_axis, evals_result['validation_1']['logloss'], label='Validación')
plt.xlabel('Iteraciones')
plt.ylabel('Log Loss')
plt.title('Curvas de Log Loss - Entrenamiento vs Validación \n (XGBoost con RandomizedSearchCV)')
plt.legend()
plt.grid(True)
plt.show()

# --- Evaluación del modelo ---
y_pred = best_model.predict(X_test)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
ap20 = average_precision_score(y_test, y_pred)

print("Métricas de rendimiento en el conjunto de prueba:")
print(f"  F1: {f1:.2f}")
print(f"  Precisión: {precision:.2f}")
print(f"  mAP@20: {ap20:.2f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

# --- Guardar el modelo entrenado ---
joblib.dump(best_model, "trained_models/xgb_model.joblib")
```

# Intento de XGBoost 3er intento con Oversampling (SMOTE - Mejor Modelo)

```python
import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import f1_score, precision_score, average_precision_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Técnicas de oversampling
from imblearn.over_sampling import SMOTE 

import xgboost as xgb
from xgboost import XGBClassifier

# Crear carpeta para guardar modelos entrenados
os.makedirs('trained_models', exist_ok=True) 

# --- Paso 1: Cargar y combinar todas las bases binarias ---
orig_folder = "new_products_matrices"

files = [os.path.join(orig_folder, f) for f in os.listdir(orig_folder)
         if f.endswith(".csv") and f.startswith("new_products_matrix_")]

list_df = []
for file in files:
    df = pd.read_csv(file, index_col=0)
    list_df.append(df)

df_all = pd.concat(list_df, axis=0)

# --- Paso 2: Definir el target ---
y = (df_all.sum(axis=1) > 0).astype(int)
X = df_all.values

print("Distribución de clases original:")
print(pd.Series(y).value_counts())

# --- Paso 3: Imputar valores faltantes usando estrategia "constant" con fill_value=0 ---
imputer = SimpleImputer(strategy='constant', fill_value=0)
X_imputed = imputer.fit_transform(X)

# --- Paso 4: Preprocesamiento: escalado y reducción de dimensionalidad (PCA) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

n_components = min(20, X_scaled.shape[1])
pca = PCA(n_components=n_components, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# --- Paso 5: Dividir en entrenamiento y prueba (70-30) ---
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.3, stratify=y, random_state=42
)

# --- Paso 6: Aplicar oversampling (SMOTE) en el conjunto de entrenamiento ---
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
print("Distribución de clases después de SMOTE:")
print(pd.Series(y_train_res).value_counts())

# --- Paso 7: Calcular scale_pos_weight para XGBoost ---
n_neg = (y_train == 0).sum()
n_pos = (y_train == 1).sum()
scale_pos_weight = n_neg / n_pos if n_pos > 0 else 1
print(f"scale_pos_weight calculado: {scale_pos_weight:.2f}")

# --- Paso 8: Ajuste de hiperparámetros con RandomizedSearchCV ---
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 1.5, 2],
    'scale_pos_weight': [scale_pos_weight]
}

clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
random_search = RandomizedSearchCV(
    clf, 
    param_distributions=param_grid, 
    scoring='f1', 
    cv=3, 
    n_iter=20, 
    n_jobs=-1, 
    random_state=42
)
random_search.fit(X_train_res, y_train_res)
best_model = random_search.best_estimator_
print("Mejores hiperparámetros encontrados:", random_search.best_params_)

# --- Paso 9: Entrenar el modelo final ---
eval_set = [(X_train_res, y_train_res), (X_test, y_test)]
best_model.fit(
    X_train_res, y_train_res, 
    eval_set=eval_set,  
    verbose=True
)

evals_result = best_model.evals_result()

# --- Paso 10: Graficar curvas de entrenamiento y validación (Logloss) ---
epochs = len(evals_result['validation_0']['logloss'])
x_axis = range(epochs)
plt.figure(figsize=(10,6))
plt.plot(x_axis, evals_result['validation_0']['logloss'], label='Entrenamiento')
plt.plot(x_axis, evals_result['validation_1']['logloss'], label='Validación')
plt.xlabel('Iteraciones')
plt.ylabel('Log Loss')
plt.title('Curvas de Log Loss - Entrenamiento vs Validación\n (XGBoost con SMOTE)')
plt.legend()
plt.grid(True)
plt.show()

# --- Paso 11: Evaluación del modelo ---
y_pred = best_model.predict(X_test)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
ap20 = average_precision_score(y_test, y_pred)

print("Métricas de rendimiento en el conjunto de prueba:")
print(f"  F1: {f1:.2f}")
print(f"  Precisión: {precision:.2f}")
print(f"  mAP@20: {ap20:.2f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

# --- Paso 12: Guardar el modelo entrenado ---
joblib.dump(best_model, "trained_models/xgb_model_all_data_oversampled.joblib")
```

# Modelo CPS con BalancedRandomForestClassifier

```python
import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import f1_score, precision_score, average_precision_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Importar un modelo cost‐sensitive: BalancedRandomForestClassifier de imblearn
from imblearn.ensemble import BalancedRandomForestClassifier

# Crear carpeta para guardar los modelos entrenados
os.makedirs('trained_models', exist_ok=True)

# --- Paso 1: Cargar y combinar todas las bases binarias ---
orig_folder = "new_products_matrices"

# Se asume que los archivos tienen nombres tipo "new_products_matrix_2007_2008.csv", etc.
files = [os.path.join(orig_folder, f) for f in os.listdir(orig_folder)
         if f.endswith(".csv") and f.startswith("new_products_matrix_")]

list_df = []
for file in files:
    df = pd.read_csv(file, index_col=0)
    list_df.append(df)

# Combinar verticalmente todas las bases
df_all = pd.concat(list_df, axis=0)

# --- Paso 2: Definir el target ---
# Se asigna 1 si la suma de la fila (productos nuevos) es > 0, 0 en caso contrario.
y = (df_all.sum(axis=1) > 0).astype(int)
X = df_all.values

print("Distribución de clases original:")
print(pd.Series(y).value_counts())

# --- Paso 3: Imputar valores faltantes ---
# Dado que la matriz es binaria, imputamos NaN con 0
imputer = SimpleImputer(strategy='constant', fill_value=0)
X_imputed = imputer.fit_transform(X)

# --- Paso 4: Preprocesamiento: escalado y reducción de dimensionalidad (PCA) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

n_components = min(20, X_scaled.shape[1])
pca = PCA(n_components=n_components, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# --- Paso 5: Dividir en entrenamiento y prueba (70-30) ---
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.3, stratify=y, random_state=42
)

# --- Paso 6: Entrenar un modelo cost‐sensitive (BalancedRandomForestClassifier) ---
# Este modelo internamente balancea las clases usando submuestreo.
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'max_features': ['sqrt', 'log2', 0.5]
}

brf = BalancedRandomForestClassifier(random_state=42, n_jobs=-1)
random_search = RandomizedSearchCV(
    brf,
    param_distributions=param_grid,
    scoring='f1',
    cv=3,
    n_iter=10,
    n_jobs=-1,
    random_state=42
)
random_search.fit(X_train, y_train)
best_model = random_search.best_estimator_
print("Mejores hiperparámetros encontrados:", random_search.best_params_)

# --- Paso 7: Evaluación del modelo ---
y_pred = best_model.predict(X_test)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
ap20 = average_precision_score(y_test, y_pred)

print("Métricas de rendimiento en el conjunto de prueba:")
print(f"  F1: {f1:.2f}")
print(f"  Precisión: {precision:.2f}")
print(f"  mAP@20: {ap20:.2f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

# --- Paso 8: Guardar el modelo entrenado ---
joblib.dump(best_model, "trained_models/brf_model_all_data_70_30_cps.joblib")
```
